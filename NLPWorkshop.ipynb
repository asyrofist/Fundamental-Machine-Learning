```
#menghitung frekuensi corpus gutenberg tertentu
from collections import Counter
cnt = Counter()
for word in nltk.corpus.gutenberg.words('shakespeare-macbeth.txt'):
  word = word.lower()
  cnt[word] += 1

# cnt['scotland']
# cnt['thane']
# cnt['blood']
cnt['duncan']
```

```
# mengurutkan berdasarkan frekeunsi
cnt.most_common()[:20]
```

```
# mengurutkan berdasarkan freqdist
import nltk
sec_a = nltk.corpus.gutenberg.words()
fd = nltk.FreqDist(sec_a)
from operator import itemgetter
swc = sorted(fd.items(), key=itemgetter(1), reverse=True)
[token for (token, freq) in swc[:20]]
```

```
# tokenizing
sentences = "Nama saya Rakha Asyrofi"
words = nltk.word_tokenize(sentences)
words
```

```
# stemming
stemmer = nltk.PorterStemmer()
verbs = ['calling', 'called', 'appears', 'appear', 'appeared']
stems = []
for verb in verbs:
  stemmed_verb = stemmer.stem(verb)
  stems.append(stemmed_verb)

# stems
# set(stems)
sorted(set(stems))
```

```
# import guttenberg collection
from nltk.corpus import gutenberg

# what corpora are in the collection
# print (gutenberg.fileids())

# import freqdist class
from nltk.probability import FreqDist

# create frequency distribution object
fd = FreqDist()

# for each token in the relevant text, increment its counter
for word in gutenberg.words('austen-persuasion.txt'):
  fd[word] += 1

# # total number of samples
# print(fd.N())

# # number of bins or unique samples
# print(fd.B())

# get a list of the top 10 words sorted by Frequency
for word in list(fd.keys()):
  print (word, fd[word])
```

```
# insugural corpus
from nltk.corpus import inaugural
inaugural.fileids()[:2]

cfd = nltk.ConditionalFreqDist(
    [(fileid[:4], target) for fileid in inaugural.fileids() for w in inaugural.words(fileid) for target in
     ['america', 'war'] if w.lower().startswith(target)])
cfd.plot()
```
